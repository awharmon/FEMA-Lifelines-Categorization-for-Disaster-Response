## Correlation analysis
> So where are we? On the one hand, we have a yelp dataset of thousands of yelp listing for the disaster zone Houston. But, it's important to understand Houston itself as a city, and fortunately zip codes provide a neat numeric organization of different city sections. Thus, the next move that made the most sense was to take Houston zip codes and gather as much additional data on them as possible, compiling them into a neat dataframe for the following analysis.

```python
fix, ax = plt.subplots(figsize=(15,15))

ax = sns.heatmap(df.corr(), ax=ax, annot=True, cmap='coolwarm')

ax.set_xticklabels(ax.xaxis.get_ticklabels(), fontsize=13)
ax.set_yticklabels(ax.yaxis.get_ticklabels(), fontsize=13);
```

```python
# Made a  scatter function
def scatter_p(fea,price):
    n_f = len(fea)
    n_row = n_f//3+1
    fig=plt.figure(figsize=(20,15))
    i = 1
    for f in fea:
        x=df[f]
        y=df[price]
        m, b = np.polyfit(x, y, 1)    
        
        ax=fig.add_subplot(n_row,3,i)
        plt.plot(x,y,'.',color='b')
        plt.plot(x, m*x + b, '-',color='r')
        plt.xlabel(f)
        plt.ylabel(price)
        i += 1
```

```python
plt.figure(figsize=(6,10))
sns.heatmap(df.corr()[['med_hh_income']].sort_values(by='med_hh_income',
                                                ascending=False), cmap='coolwarm',
           vmin=-1,
           vmax=1,
           annot=True)
plt.title('Correlations for median household income \n');
```

```python
plt.figure(figsize=(6,10))
sns.heatmap(df.corr()[['zip_code']].sort_values(by='zip_code',
                                                ascending=False), cmap='coolwarm',
           vmin=-1,
           vmax=1,
           annot=True)
plt.title('Correlations or lack thereof for zip code \n');
```

### Some quick and dirty inferences so far:
1. It seems like median household income is a pretty strong predictor for zillow's housing index. I think it's reasonable to assume that they factor in income a lot when doing their time series analysis as well. 
2. Zip code doesn't really correlate well with anything that isn't also spacial, unfortunately. At best it has a small connection to household income.
3. It would be very easy to fall down a rabbit hole of visualizing housing data and the 2008 financial crisis and things like that, but that's not particularly relevant or a good use of our time when looking at natural disasters.
