```python
path = "./HARVEY_BusinessList_10-16-2017.txt"
```

```python
damaged_biz = pd.read_csv(path,delimiter= "|" )
```

```python
damaged_biz.shape
```

```python
damaged_biz.head()
```

```python
damaged_biz["FULLZIP"] = damaged_biz["FULLZIP"].map(lambda x: str(x))
```

```python
ammended_zc = []
for each in damaged_biz["FULLZIP"]:
#     print(each)
    if len(each) > 5:
        new_each = int(each[:-4])
        ammended_zc.append(new_each)
    else:
        ammended_zc.append(each)
```

```python
len(ammended_zc)
```

```python
damaged_biz["new_zc"] = ammended_zc
```

```python
damaged_biz_dict = damaged_biz.groupby("new_zc").count()["NAME"].to_dict()
```
