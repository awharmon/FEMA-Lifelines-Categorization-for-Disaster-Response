## Safety and Security  (ss)

```python
businesses_ss = []
offset = 0

for each in saftey_and_securiy:
        print(each)
        offset = 0
        for i in range(20):
            print(i)
            old_len = len(businesses_ss)
            params = {"location": "houston", "categories": each, "limit":50, "offset": offset}
            url = url
            req = requests.request("GET", url, data=payload, headers=headers, params=params)
            if req.status_code != 200: 
                print(json.loads(req.text))
                print(each)
                break
            else:
                parsed = json.loads(req.text)
                businesses_ss.extend(parsed["businesses"])
                print(len(businesses_ss))
                new_len = len(businesses_ss)
# #                 time.sleep(1)  
            offset = offset + 50
            if old_len == new_len:
                break
# # Thanks to Joe Darby for his contrabutions 
```

```python
all_data_ss = []
for each in businesses_ss:
    values = []
    try:
        price = each["price"]
    except:
        price = "NAN" 
    for ix in each["categories"]:
        value = list(ix.values())
        values.append(value)
    data_dict_ss = {
        'id': each["id"], 
        'alias': each["alias"], 
        'name':each["name"], 
        'image_url':each["image_url"], 
        'is_closed':each["is_closed"], 
        'url':each["url"], 
        'review_count':each["review_count"], 
        'categories': values,
        'rating':each["rating"], 
        'latitude': each["coordinates"].get("latitude"), 
        'longitude':each["coordinates"].get("longitude"), 
        'transactions': each["transactions"], 
        'price': price, 
        'address1':each["location"].get("address1"), 
        'address2':each["location"].get("address2"), 
        'address3':each["location"].get("address3"), 
        'city': each["location"].get('city'), 
        'zip_code': each["location"].get('zip_code'), 
        'country':each["location"].get('country'), 
        'state':each["location"].get('state'), 
        'display_address':each["location"].get('display_address'), 
        'phone':each["phone"], 
        'display_phone':each["display_phone"], 
        'distance':each["distance"]
    }
    all_data_ss.append(data_dict_ss)
```

```python
df_ss = pd.DataFrame(all_data_ss)
df_ss.shape
```

```python
df_ss["fema_l1"] = 1
```

```python
df_ss.head()
```

```python
path = "../Datasets/fema_ll_ss_houston.csv"
data_frame = df_ss.to_csv(path, index = False)
```
